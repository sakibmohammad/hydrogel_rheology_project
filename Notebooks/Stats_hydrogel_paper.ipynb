Feature Importance for Forward regression and correlation matrix

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor


df = pd.read_csv('/content/drive/MyDrive/Data for ML Project (2).csv')
#df.info()

#df.dropna(axis=1, inplace=True)
'''
print(f"\n\n\nData information: \n{df.info()}")
print(f"\n\n\nData description: \n{df.describe()}")
print(f"\n\n\nMean of every column: \n{df.mean()}")
print(f"\n\n\nMedian of every column: \n{df.median()}")
print(f"\n\n\nVariance of every column: \n{df.var()}")
print(f"\n\n\nStd of every column: \n{df.std()}")
print(f"\n\n\nUnique values in every column: \n{df.nunique()}")
'''

out_vars = ['Storage modulus (Pa)', 'Loss modulus (Pa)' ]
plt.figure(figsize=(10, 20))

for idx, var in enumerate(out_vars, 1):
#df = df.values
  X = df.drop(['Storage modulus (Pa)', 'Loss modulus (Pa)'], axis = 1)

  y = df[var]

  model = XGBRegressor()
  model.fit(X, y)
  importances = pd.DataFrame(data={
    'Attribute': X.columns,
    'Importance': model.feature_importances_
    })
  importances = importances.sort_values(by='Importance', ascending=False)

  plt.subplot(2, 1, idx)

  plt.bar(x=importances['Attribute'], height=importances['Importance'], color ='orange', linewidth = 2.0, edgecolor='black')
  ax = plt.gca()
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  ax.yaxis.set_ticks_position('left')
  ax.xaxis.set_ticks_position('bottom')
  ax.xaxis.set_tick_params(width = 2.0)
  ax.yaxis.set_tick_params(width = 2.0)

  plt.title(f'Feature Importance for {var} Prediction', fontsize = 16, fontweight = 'bold', fontname='Liberation Sans')
  plt.xticks(rotation='vertical', fontsize = 16, fontweight = 'bold', fontname='Liberation Sans')
  plt.yticks(fontsize = 16, fontweight = 'bold', fontname='Liberation Sans')

plt.tight_layout()
plt.show()


corrmat = df.corr()
plt.figure(figsize=(10, 6))
hm = sns.heatmap(corrmat,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 linewidth = 1.0,
                 linecolor= 'black',
                 annot_kws={'size': 12,
                            'fontweight':'bold',
                            'fontname':'Liberation Sans'},
                 yticklabels=corrmat.columns,
                 xticklabels=corrmat.columns
                 )
plt.show()
```

Pearson's correlation among features and g', g".

```python
from scipy import stats
df = pd.read_csv('/content/drive/MyDrive/Data for ML Project (2).csv')

print("Acrylamide Conc. %\n")
for i in df.columns:
  if i != 'Storage modulus (Pa)' and i!='Loss modulus (Pa)':
    r, p = stats.pearsonr(df[i], df['Storage modulus (Pa)'])
    print(f"Feature: {i}, Correlation: {r}, Significant (p<0.05): {p<0.05}")

print("Acrylamide Conc. %\n")
for i in df.columns:
  if i != 'Storage modulus (Pa)' and i!='Loss modulus (Pa)':
    r, p = stats.pearsonr(df[i], df['Loss modulus (Pa)'])
    print(f"Feature: {i}, Correlation: {r}, Significant (p<0.05): {p<0.05}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor

df = pd.read_csv('/content/drive/MyDrive/Hydrogel_ML_data.csv')
#df.info()

# drop any rows with missing values
#df.dropna(axis=1, inplace=True)
'''
print(f"\n\n\nData information: \n{df.info()}")
print(f"\n\n\nData description: \n{df.describe()}")
print(f"\n\n\nMean of every column: \n{df.mean()}")
print(f"\n\n\nMedian of every column: \n{df.median()}")
print(f"\n\n\nVariance of every column: \n{df.var()}")
print(f"\n\n\nStd of every column: \n{df.std()}")
print(f"\n\n\nUnique values in every column: \n{df.nunique()}")
'''

out_vars = ['Acrylamide Conc. %', 'Bis-acrylamide conc %', 'Photo-initiator conc. %', 'Layer Height. (micron)',
          'Bottom Layer exposure time (s) ', 'Exposure time (s)']

plt.figure(figsize=(10, 20))
for idx, var in enumerate(out_vars, 1):
#df = df.values
  X = df.drop(['Acrylamide Conc. %', 'Bis-acrylamide conc %', 'Photo-initiator conc. %', 'Layer Height. (micron)',
          'Bottom Layer exposure time (s) ', 'Exposure time (s)'], axis = 1)

  y = df[var]



  model = XGBRegressor()
  model.fit(X, y)
  importances = pd.DataFrame(data={
    'Attribute': X.columns,
    'Importance': model.feature_importances_
  })
  importances = importances.sort_values(by='Importance', ascending=False)

  plt.subplot(3, 3, idx)
  plt.bar(x=importances['Attribute'], height=importances['Importance'], color ='steelblue', linewidth = 2.0, edgecolor='black')
  ax = plt.gca()
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  ax.yaxis.set_ticks_position('left')
  ax.xaxis.set_ticks_position('bottom')
  ax.xaxis.set_tick_params(width = 2.0)
  ax.yaxis.set_tick_params(width = 2.0)

  plt.title(f'Feature Importance for {var} Prediction', fontsize = 7, fontweight = 'bold', fontname='Liberation Sans')
  plt.xticks(rotation='vertical', fontsize = 14, fontweight = 'bold', fontname='Liberation Sans')
  plt.yticks(fontsize = 14, fontweight = 'bold', fontname='Liberation Sans')

plt.tight_layout()
plt.show()
```
